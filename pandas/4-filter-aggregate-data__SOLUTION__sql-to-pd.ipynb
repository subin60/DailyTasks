## 6. Which category of books has had more year-to-date sales, "business" or "popular_comp"? You can write two queries to answer this question.

````python
# Tania

bs = titles[titles["type"] == "business"]
bs_ytd = bs.groupby("type").agg({"ytd_sales":"sum"})

pc = titles[titles["type"] == "popular_comp"]
pc_ytd = pc.groupby("type").agg({"ytd_sales":"sum"})

bs_ytd["ytd_sales"][0] > pc_ytd["ytd_sales"][0]
````

````python
bs_ytd
````

````python
pc_ytd
````

````python
# Therese

bs_ytd = titles[titles["type"] == "business"][["ytd_sales"]].agg("sum")
pc_ytd = titles[titles["type"] == "popular_comp"][["ytd_sales"]].agg("sum")

bs_ytd > pc_ytd
````

### Hint:

In SQL the syntax is:

```sql
SELECT SUM(ytd_sales), type 
FROM titles
WHERE type = "business";
```

### Hint:

In SQL the syntax is:

```sql
SELECT SUM(ytd_sales), type 
FROM titles
WHERE type = "popular_comp";
```

## 9. How many authors live in either San Jose or Salt Lake City
